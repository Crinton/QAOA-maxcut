```python
import numpy as np                                          # 导入numpy库并简写为np
from mindquantum.core.gates import X, Y, Z, H, RX, RY, RZ,ZZ   # 导入量子门H, X, Y, Z, RX, RY, RZ
from mindquantum.simulator import Simulator    # 从mindquantum.simulator中导入Simulator类
from mindquantum.core.circuit import Circuit   


def Rz(t:float) ->np.array:
    return np.cos(t/2)*np.eye(2) - 1j*np.sin(t/2)*np.array([[1,0],
                                                      [0,-1]])
def Rx(t:float) ->np.array:
    return np.cos(t/2)*np.eye(2) - 1j*np.sin(t/2)*np.array([[0,1],
                                                      [1,0]])
def Ry(t:float) ->np.array:
    return np.cos(t/2)*np.eye(2) - 1j*np.sin(t/2)*np.array([[0,-1j],
                                                      [1j,-1]])

def Rzz(t:float) -> np.array:
    z = np.array([[1,0],
                  [0,-1]])
    return np.cos(t/2)*np.eye(4) - 1j*np.sin(t/2)*np.kron(z,z)
def Rzzz(t:float) -> np.array:
    z = np.array([[1,0],
                  [0,-1]])
    return np.cos(t/2)*np.eye(8) - 1j*np.sin(t/2)*np.kron(np.kron(z,z),z)
```

```python
beta_lst = np.linspace(0,2*np.pi,200)
gamma_lst = np.linspace(0,2*np.pi,200)

```

```python
E_max = 0
para_dict = {"gamma":0,"beta":0}
pbar = tqdm(total=len(beta_lst)*len(gamma_lst))
sim = Simulator('mqvector', 2)
C = QubitOperator('Z0 Z1') + QubitOperator('Z1') + QubitOperator('Z0')
H_c = Hamiltonian(C)
for i,beta in enumerate(beta_lst):
    for j,gamma in enumerate(gamma_lst):
        sim.reset()
        circuit = Circuit()
        circuit += H.on(0)
        circuit += H.on(1)
        circuit.barrier()
        circuit += ZZ('gamma_2').on((0,1))
        circuit += RZ('gamma_1').on(0)
        circuit += RZ('gamma_1').on(1)
        circuit.barrier()
        circuit += RX('beta').on(0)
        circuit += RX('beta').on(1)
        #circuit.measure_all()      
        sim.apply_circuit(circuit,pr = {"gamma_1":gamma,"gamma_2":gamma/2,"beta":beta})
        #sim.sampling(circuit)
        E = sim.get_expectation(H_c).real                       # 为线路中所有比特添加测量门
        #print("E_Max = {}, E = {}".format(E_max,E))
        if E > E_max:
            E_max = E
            para_dict["gamma"] = gamma
            para_dict["beta"] = beta
        else:
            pass
        pbar.update(1)
#sim.sampling(circuit, shots=1000).svg()
```
