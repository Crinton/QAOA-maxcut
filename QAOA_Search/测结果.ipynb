```python
import numpy as np  
import pandas as pd
import itertools
```

```python
def func_11(z1,z2,z3,z4):
    C_1 = z1
    C_2 = z3
    C_3 = -z4
    C_4 = -z1*z2*z3*z4
    return C_1 + C_2 + C_3 + C_4
Vars_lst = itertools.product([-1,1],[-1,1],[-1,1],[-1,1])
obj_max = 0
para_max = []
for z1,z2,z3,z4 in Vars_lst:
    obj = func_11(z1,z2,z3,z4)
    #if obj >= obj_max:
    #    obj_max = obj
    #   print("z1: {}, z2: {}, z3: {}, z4: {},".format(z1,z2,z3,z4))
    para_max.append((z1,z2,z3,z4,obj))
df = pd.DataFrame(para_max,columns=["z1","z2","z3","z4","obj"])
df.sort_values("obj",ascending=False).to_csv("test_11.csv")
```
